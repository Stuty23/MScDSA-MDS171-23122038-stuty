### Matrix Multiplication

```python
# Initialize the dictionary to store matrices
matrix_dict = {
    'matrix 1': [],
    'matrix 2': [],
    'result matrix': []
}

# Function to collect a matrix from the user
def collect_matrix(matrix_name):
    matrix = []
    rows = int(input(f"Enter the number of rows for {matrix_name}: "))
    columns = int(input(f"Enter the number of columns for {matrix_name}: "))
    
    print(f"Enter the elements of {matrix_name} row-wise:")
    for i in range(rows):
        row = list(map(float, input(f"Enter row {i + 1}: ").split()))
        if len(row) != columns:
            print(f"Error: Row {i + 1} should have {columns} elements.")
            return None
        matrix.append(row)
    
    matrix_dict[matrix_name] = matrix

# Function to perform matrix multiplication
def multiply_matrices():
    matrix1 = matrix_dict['matrix 1']
    matrix2 = matrix_dict['matrix 2']
    
    if len(matrix1[0]) != len(matrix2):
        print("Error: The number of columns in matrix 1 should be equal to the number of rows in matrix 2.")
        return
    
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            product = 0
            for k in range(len(matrix2)):
                product += matrix1[i][k] * matrix2[k][j]
            row.append(product)
        result.append(row)
    
    matrix_dict['result matrix'] = result

# Function to display the result in a file
def export_to_file(filename):
    with open(filename, 'w') as file:
        result_matrix = matrix_dict['result matrix']
        for row in result_matrix:
            file.write(' '.join(map(str, row)) + '\n')
    print(f"Result matrix has been exported to {filename}.")

# Main menu
while True:
    print("\nMain Menu:")
    print("1. Enter Matrix 1")
    print("2. Enter Matrix 2")
    print("3. Calculate")
    print("4. Export to File")
    print("5. Quit")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        collect_matrix('matrix 1')
    elif choice == '2':
        collect_matrix('matrix 2')
    elif choice == '3':
        if not matrix_dict['matrix 1'] or not matrix_dict['matrix 2']:
            print("Error: Please enter both Matrix 1 and Matrix 2.")
        else:
            multiply_matrices()
            print("Matrix multiplication completed.")
    elif choice == '4':
        if not matrix_dict['result matrix']:
            print("Error: Please calculate the result matrix first.")
        else:
            filename = input("Enter the filename to export the result matrix: ")
            export_to_file(filename)
    elif choice == '5':
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
